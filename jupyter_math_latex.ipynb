#### 空格

在LaTeX中，符号之间的空格会被自动移除，所以 \int f(x) dx  显示为 
$$ \int f(x) dx $$

- 通过 `\,` 或 `\:` 或 `\;` 添加空格， 其空格宽度分别为从小到大
    - `$ \int f(x) \, dx $` 显示为 $$ \int f(x) \, dx $$
    - `$ \int f(x) \: dx $` 显示为 $$ \int f(x) \: dx $$
    - `$ \int f(x) \; dx $` 显示为 $$ \int f(x) \; dx $$

- 更大的间隔，可以使用 \quad 和 \qquad 。第一个为 1em 的距离，第二个为 2em 的距离；
    - `$\int \quad f(x)$` 表示 $$\int \quad f(x)$$
    - `$\int \qquad f(x)$` 表示 $$\int \qquad f(x)$$

#### 等式对齐

- 通过 `\\` 断开两个或多个等式，可实现等式中部对齐，例如：
`$ a_1=b_1+c_1 \\ a_2=b_2+c_2+d_2 \\ a_3=b_3+c_3 $ ` 表示

$$
a_1=b_1+c_1 \\
a_2=b_2+c_2+d_2 \\
a_3=b_3+c_3
$$

- 如果希望是左对齐，可以是：
`$ \begin{align} a_1&=b_1+c_1 \\ a_2&=b_2+c_2+d_2 \\ a_3&=b_3+c_3 \end{align} $` 表示

$ \begin{align} a_1&=b_1+c_1 \\ a_2&=b_2+c_2+d_2 \\ a_3&=b_3+c_3 \end{align} $

- 如果需要对方程进行编号，可以 使用 \tag ，例如：
`$ \begin{align} a_1&=b_1+c_1 \tag{公式1} \\ a_2&=b_2+c_2+d_2 \tag{公式2} \\ a_3&=b_3+c_3 \tag{公式3} \end{align} $`

$ \begin{align} a_1&=b_1+c_1 \tag{公式1}\\ a_2&=b_2+c_2+d_2 \tag{公式2}\\ a_3&=b_3+c_3 \tag{公式3} \end{align} $

- `$ \begin{align}\begin{cases} a_1&=b_1+c_1  \\ a_2&=b_2+c_2+d_2  \cr a_3&=b_3+c_3  \end{cases}\end{align} $` 表示

$
\begin{align}\begin{cases}
a_1&=b_1+c_1  \\ a_2&=b_2+c_2+d_2  \cr a_3&=b_3+c_3 
\end{cases}\end{align}
$
